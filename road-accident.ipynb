```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
data=pd.read_csv("/kaggle/input/road-accident-data-2020-india/df.csv")
```

```python
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
data1=data.fillna(0.0)
```

```python
data1.info()
```

```python
data1["Million Plus Cities"].value_counts()
```

```python
Cause_counts = data1['Cause category'].str.split(',').explode().value_counts()


plt.figure(figsize=(10,11))
sns.barplot(x=Cause_counts.values, y=Cause_counts.index, palette="viridis")
plt.title("Cause catagory")
plt.xlabel("Count")
plt.ylabel("Couse catagory")
plt.show()
```

```python
outcome_counts = data1['Outcome of Incident'].str.split(',').explode().value_counts()


plt.figure(figsize=(10,11))
sns.barplot(x=outcome_counts.values, y=outcome_counts.index, palette="plasma")
plt.title("outcome")
plt.xlabel("Count")
plt.ylabel("outcome_counts")
plt.show()
```

```python
cause_Subcategory_counts = data1['Cause Subcategory'].str.split(',').explode().value_counts()


plt.figure(figsize=(10,11))
sns.barplot(x=cause_Subcategory_counts.values, y=cause_Subcategory_counts.index, palette="magma")
plt.title("cause_Subcategory")
plt.xlabel("Count")
plt.ylabel("cause_Subcategory")
plt.show()
```
